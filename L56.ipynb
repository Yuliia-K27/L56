```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as clr
```

```python
im = cv2.imread('kitty.jpeg')
```

```python
plt.imshow(im)
plt.axis('off')
plt.show
```

```python
im_RGB = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(im_RGB)
plt.show
```

### Декомпозиція RGB

#### Декомпозиція RGB по R

```python
R = im_RGB.copy()
R[:,:,(1,2)] = 0
plt.axis('off')
plt.imshow(R)
plt.show
```

#### Декомпозиція RGB по G

```python
G = im_RGB.copy()
G[:,:,(0,2)] = 0
plt.axis('off')
plt.imshow(G)
plt.show
```

#### Декомпозиція RGB по B

```python
B = im_RGB.copy()
B[:,:,(0,1)] = 0
plt.axis('off')
plt.imshow(B)
plt.show()
```

**Зберемо декомпозиційні зображення в єдине**

```python
new_RGB = R + G + B
plt.axis('off')
plt.imshow(new_RGB)
plt.show()
```

### Конвертування RGB у HSV

```python
im_HSV = cv2.cvtColor(im_RGB, cv2.COLOR_BGR2HSV)
plt.axis('off')
plt.imshow(im_HSV)
plt.show()
```

### Декомпозиція HSV

#### Декомпозиція HSV по H

```python
H=im_HSV.copy()
H[:,:,(1,2)] = 255
plt.axis('off')
plt.imshow(H)
plt.show()
```

```python
H=im_HSV.copy()
H[:,:,(1,2)] = 0
plt.axis('off')
plt.imshow(H)
plt.show()
```

#### Декомпозиція HSV по S

```python
S=im_HSV.copy()
S[:,:, 0] = 179
S[:,:, 2] = 255
plt.axis('off')
plt.imshow(S)
plt.show()
```

#### Декомпозиція HSV по V

```python
V=im_HSV.copy()
V[:,:, 0] = 179
V[:,:, 1] = 0
plt.axis('off')
plt.imshow(V)
plt.show()
```

**Зберемо декомпозиційні зображення в єдине**

```python
new_HSV = H + S + V
plt.axis('off')
plt.imshow(new_HSV)
plt.show()
```

### Конвертування RGB у CIE Lab

```python
im_Lab = cv2.cvtColor(im_RGB, cv2.COLOR_BGR2LAB)
plt.axis('off')
plt.imshow(im_Lab)
plt.show()
```

### Декомпозиція CIE Lab

#### Декомпозиція CIE Lab по L

```python
L=im_Lab.copy()
L[:,:,(1,2)] = 0
L_RGB = cv2.cvtColor(L, cv2.COLOR_LAB2RGB)
L_GRAY = cv2.cvtColor(L_RGB, cv2.COLOR_RGB2GRAY)
plt.axis('off')
plt.imshow(L_GRAY, cmap = 'gray' )
plt.show()
```

#### Декомпозиція CIE Lab по a

```python
a = im_Lab.copy()
a[:, :, 0] = 127
a[:, :, 0] = 0
a_RGB = cv2.cvtColor(a, cv2.COLOR_LAB2RGB)
a_GRAY = cv2.cvtColor(a_RGB, cv2.COLOR_RGB2GRAY)
```

```python
cmap_a = clr.LinearSegmentedColormap.from_list('custom blue', ['Red', 'Gray', 'Green'], N = 255)
plt.axis('off')
plt.imshow(a_GRAY, cmap = cmap_a )
plt.show()
```

#### Декомпозиція CIELab по b

```python
b=im_Lab.copy()
b[:,:,0] = 127
b[:,:, 1] = 0
b_RGB = cv2.cvtColor(b, cv2.COLOR_LAB2RGB)
b_GRAY = cv2.cvtColor(b_RGB, cv2.COLOR_RGB2GRAY)
```

```python
cmap_b = clr.LinearSegmentedColormap.from_list('custom blue', ['Yellow', 'Gray', 'Green'], N = 255)
plt.axis('off')
plt.imshow(b_GRAY, cmap = cmap_b)
plt.show()
```

**Зберемо декомпозиційні зображення в єдине**

```python
new_Lab = L + a + b
plt.axis('off')
plt.imshow(new_RGB)
plt.show()
```
